```python
from langchain_neo4j import Neo4jGraph


from configuration.config import *

load_dotenv(ROOT_PATH / '.env')

print(os.getenv("HF_ENDPOINT"))
```

```python
graph = Neo4jGraph(
    url=NEO4J_CONFIG['uri'],
    username=NEO4J_CONFIG['auth'][0],
    password=NEO4J_CONFIG['auth'][1]
)
```

```python
graph.query("MATCH(N) RETURN N")
```

```python
from langchain_neo4j import GraphCypherQAChain
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model='qwen-plus',
    model_provider='openai',
    api_key=os.getenv('QWEN_API_KEY'),
    base_url=os.getenv("QWEN_BASE_URL"),
    temperature=0
)

model1 = init_chat_model(
    model='deepseek-chat',
    model_provider='openai',
    api_key=os.getenv('DEEP_SEEK_API_KEY'),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0
)
DeepSeek = init_chat_model(
    model='deepseek-ai/DeepSeek-V3.2',
    model_provider='openai',
    api_key=os.getenv('MODEL_SCOPE_API_KEY'),
    base_url=os.getenv("MODEL_SCOPE_BASE_URL"),
    temperature=0
)
chain = GraphCypherQAChain.from_llm(
    # verbose：显示过程
    graph=graph, llm=DeepSeek, verbose=True, validate_cypher=True,
    allow_dangerous_requests=True
)
```

```python
chain.invoke({"query": "苹果有哪些产品？"})
```
